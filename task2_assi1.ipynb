```python
import math
```

```python
def checkPrime(num):
    sq=math.sqrt(num)
    sq=math.ceil(sq)
    count=0
    for i in range (2,sq+1):
       if num%i==0:
           count+=1
        
    if(count==0 or num==2):
        return True
    else:
        return False
```
